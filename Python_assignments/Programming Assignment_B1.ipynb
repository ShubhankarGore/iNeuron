2.	Write a Python program to do arithmetical operations addition and division.?

```python
a = int(input("Enter a number: "))
b = int(input("Enter a number: "))
o = int(input('Enter 1 for adding two numbers & enter 2 for divinding them: '))

if o == 1:
    c = a + b 
    print("addition of given nuimber is "+ str(c))
elif o == 2 and a>0 and b>0:
    c = a / b 
    print("Division of given number is: "+ str(c))
else:
    print("invalid input.")
```

3.	Write a Python program to find the area of a triangle?

```python
x = int(input("Enter the height of triangle: "))
y = int(input("Enter the base of triangle: "))

A = 1/2 * x * y
print("Area of Triangle is: " + str(A))
```

4.	Write a Python program to swap two variables?

```python
p = int(input("Enter first number: "))
q = int(input("Enter second number: "))

s = p
p = q
q = s 

print ("first number is: "+ str(p))
print ("second number is: "+ str(q))
```
