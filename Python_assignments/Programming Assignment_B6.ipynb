1.	Write a Python Program to Display Fibonacci Sequence Using Recursion?

```python
t = int(input("Enter the number of term you want: "))

def rec_fibo(n):
    if n <= 1:
        return n
    else:
        return (rec_fibo(n-1) + rec_fibo(n-2))

if t <= 0:  
   print("Plese enter a positive integer")
else:
    for i in range(t):
        print(rec_fibo(i))
    
    
```

2.	Write a Python Program to Find Factorial of Number Using Recursion? 

```python
f = int(input("Enter the number to find it's factorial: "))

def fact(num):
    if num == 1:
        return num
    else:
        return(num * fact(num-1))
    
    
if f < 0:
    print("Factorial does not exist fornegative numbers.")
elif f == 0:
    print("Factorial of 0 is 1.")
else:
    print("the fcatorial of", f ,"is", fact(f))
```

5.	Write a Python Program for cube sum of first n natural numbers?

```python
n = int(input("Enter the number of term you want: "))
sum = 0

for i in range(1, n+1):
    sum = sum + i*i*i
print(sum)
```
