2.	Write a Python program to convert Celsius to Fahrenheit?

```python
C = int(input("Enter the Temperature in Celsius: "))

#(1°C × 9/5) + 32 = 33.8°F
F = (C * 9/5) + 32


print(str(C)+ " Celsius are equal to "+ str(F) +" Fahrenheit.")
```

4.	Write a Python program to solve quadratic equation?

```python
# Solve the quadratic equation ax**2 + bx + c = 0
import cmath

a = int(input("Enter A of the above Quadratic equation: "))
b = int(input("Enter B of the above Quadratic equation: "))
c = int(input("Enter C of the above Quadratic equation: "))
#a = 1, b = 5, c = 6

# calculate the discriminant
d = (b**2) - (4*a*c)

# find two solutions
if d >= 0:
    sol1 = (-b-cmath.sqrt(d))/(2*a)
    sol2 = (-b+cmath.sqrt(d))/(2*a)
    print('The solution are {0} and {1}'.format(sol1,sol2))
else:
    print("equation with given numbers has only imaginary solution.")

```
